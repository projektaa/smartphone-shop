```python

import pandas as pd
import numpy as np

# Read dataset
dataset= pd.read_csv('/content/topSubscribed.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
# Info column type
dataset.info()
```

```python
# Clean data

dataset["Subscribers"] = dataset["Subscribers"].apply(lambda x:int(x.replace(",","")))
dataset["Video Views"] = dataset["Video Views"].apply(lambda x:int(x.replace(",","")))
dataset["Video Count"] = dataset["Video Count"].apply(lambda x:int(x.replace(",","")))
```

```python
dataset.info()
```

```python
# Find Average
dataset['Average Views per Video'] = dataset['Video Views'] / dataset['Video Count']

# Sort the DataFrame by average views per video in descending order
dataset_sorted = dataset.sort_values(by='Average Views per Video', ascending=False)

# Select the top 5 channels
top_5_channels = dataset_sorted.head(5)

```
